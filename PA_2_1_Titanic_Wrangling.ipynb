```python
import numpy as np
import pandas as pd

data_dir = "https://dlsun.github.io/pods/data/"
df_titanic = pd.read_csv(data_dir + "titanic.csv")
```

## 1. Filter the data to include passengers only. Calculate the joint distribution (cross-tab) between a passenger's class and where they embarked.

```python
df_passengers = df_titanic[(df_titanic["class"] == "1st") | (df_titanic["class"] == "2nd") | (df_titanic["class"] == "3rd")]

pd.crosstab(df_passengers["class"], df_titanic["embarked"])
```

## 3. Make a visualization showing the distribution of a passenger's class, given where they embarked.

Discuss the pros and cons of using this visualization versus the distributions you calculated before, to answer the previous questions.

```python
import matplotlib.pyplot as plt

crosstab = pd.crosstab(df_passengers['class'], df_passengers['embarked'], normalize='columns')

crosstab.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Passenger Class')
plt.ylabel('Proportion')

#This visualization is good because it easily allows you to see trends within the data via the color-coded
  #bars and heights. It is easy to see how each bar compares to the others. This visualization is not the
  #best when it comes to noting the exact numberical heights of the bars. The y-axis does not have every
  #specific possible proportion, so the reader will have to estimate the exact heights of each of the bars.
```
